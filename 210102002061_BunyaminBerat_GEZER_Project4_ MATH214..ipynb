```python


def numpy_cholesky(A):
    # NumPy Cholesky faktörizasyonu
    numpy = np.linalg.cholesky(A)
    return numpy

def scipy_cholesky(A):
    # SciPy Cholesky faktörizasyonu
    scipy =cholesky(A, lower=True)
    return scipy

# A matrisini tanımla
A = np.array([
    [4, 2, 2],
    [2, 6, 2],
    [2, 2, 5]
], )

# NumPy Cholesky faktörizasyonu
numpy_result = numpy_cholesky(A)

# SciPy Cholesky faktörizasyonu
scipy_result = scipy_cholesky(A)

# Sonuçları yazdır
print("NumPy Cholesky Faktörizasyonu:\n", numpy_result)
print("\nSciPy Cholesky Faktörizasyonu:\n", scipy_result)

# Karşılaştırma  # all close aynı olup olmadığını kontrol eder. 
if np.allclose(numpy_result, scipy_result):
    print("\nNumPy and  SciPy Cholesky result is the same !")
else:
    print("\nNumPy and  SciPy Cholesky result is not  the same!")
```

The Cholesky factorization functions in the NumPy and SciPy libraries are theoretically expected to produce the same results, 
but in practice, the two results may not be exactly the same due to slight differences in numerical calculations.


The cholesky function does upper triangle Cholesky factorization by default. That is, it returns the Cholesky factorization of the given matrix as a lower triangular matrix L. 
However, if you set the lower parameters to True, the function returns the lower triangular of the matrix L. In this case, the scipy variable can be used to perform the lower triangular Cholesky factorization of the matrix. 
The main difference between these two cases is whether the length matrix is ​​upper or lower triangular.
If you want to get results compatible with numpy.linalg.cholesky, you can use the scipy.linalg.cholesky function without using the lower parameter or set the lower parameter to False.
